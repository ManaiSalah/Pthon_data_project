# Classes

```python
class BaseSalary:
  def __init__(obj,base_salary,bonus_rate=0.1,symbol="$"):
    obj.base_salary=base_salary
    obj.bonus_rate=bonus_rate
    obj.symbol=symbol
    obj.total_salary=base_salary*(1+bonus_rate)
    obj.bonus= obj.total_salary-obj.base_salary

  def __repr__(obj):
    return f'{obj.symbol}{obj.base_salary:,.0f}'

  def calculate_salary(obj):
   return  f'{obj.symbol}{obj.total_salary:,.0f}'

   def calculate_bonus(obj):
     return  f'{obj.symbol}{obj.bonus:,.0f}'
```

```python
salary=BaseSalary(1000)
salary.symbol
```
