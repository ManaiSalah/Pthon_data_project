```python
import pandas as pd
import matplotlib.pyplot as plt 
from datasets import load_dataset
import ast

#loading Data
dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()

#Data Cleanup 
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
df['job_skills']=df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

## Excercise Investigate Trending Skills

```python
#extracting the data for only Data Analyst jobs
df_DA=df[df['job_title_short']=='Data Analyst'].copy()
df_DA
```

```python
#Extracting the month from job_posted_date 
df_DA['job_posted_month_no']=df_DA['job_posted_date'].dt.month
```

```python
df_DA
```

```python
df_DA_explode=df_DA.explode('job_skills')

df_DA_explode
```

```python
#the parameter fill_value used to specify the value on Nan
df_DA_pivot=df_DA_explode.pivot_table(index='job_posted_month_no',columns='job_skills',aggfunc='size',fill_value=0)
```

```python
df_DA_pivot.loc['Total']=df_DA_pivot.sum()
df_DA_pivot
```

```python
df_DA_pivot.loc['Total'].sort_values(ascending=False).index
```

```python
df_DA_pivot=df_DA_pivot[df_DA_pivot.loc['Total'].sort_values(ascending=False).index]

```

```python
df_DA_pivot=df_DA_pivot.drop('Total')
```

```python
df_DA_pivot
```

```python
df_DA_pivot.iloc[:,:5].plot(
    kind='line',
    linewidth=4,  #the widh of the line
    linestyle=":" ,#options: "---",':',"-."
    colormap='plasma', #options= 'plasma','inferno','magma','viridis'
    marker='o',
    markersize=5,
    figsize=(10,7) #figure size 

   



                            
                            
)
plt.title('Top 5 skills for Data Analyst per Month')
plt.ylabel('Count')
plt.xlabel('')
plt.show()
#new parameters
```
